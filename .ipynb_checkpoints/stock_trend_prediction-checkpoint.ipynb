```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader as pdr
import yfinance as yf
```

```python
# Define date range
start = "2014-01-01"
end = "2024-12-31"

# Fetch stock data
df = yf.download("AAPL", start=start, end=end)

# Display the first few rows
df.head()
```

```python
df.tail()
```

```python
# Flatten multi-level columns and reset index
df.columns = df.columns.get_level_values(0)  # Keep only the first level (Price, Close, etc.)
df = df.reset_index()  # Make 'Date' a normal column instead of index

# Optional: Rename the 'Price' column if it's redundant
if 'Price' in df.columns:
    df = df.drop(columns=['Price'])

# Show the clean result
df.head()
```

```python
# Drop unnecessary columns
df.drop(columns=['Price', 'index'], inplace=True, errors='ignore')

# Display the cleaned dataset
df.head()
```

```python
#splitting data into training and testing
data_training = pd.DataFrame(df['Close'][0:int(len(df)*0.70)])
data_testing = pd.DataFrame(df['Close'][int(len(df)*0.70):int(len(df))])

print(data_training.shape)
print(data_testing.shape)
```

```python
data_training.head()
```

```python
data_testing.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler(feature_range=(0,1))
```

```python
data_training_array = scalar.fit_transform(data_training)
data_training_array
```

```python
data_training_array.shape
```

```python
x_train = []
y_train = []

for i in range(100, data_training_array.shape[0]):
    x_train.append(data_training_array[i-100: i])
    y_train.append(data_training_array[i, 0])
    
x_train, y_train = np.array(x_train), np.array(y_train)
```

```python
x_train.shape
```

```python
#ML Model
from keras.layers import Dense, Dropout, LSTM
from keras.models import Sequential
```

```python
model = Sequential()
model.add(LSTM(units = 50, activation='relu', return_sequences=True,
              input_shape = (x_train.shape[1],1)))
model.add(Dropout(0.2))


model.add(LSTM(units = 60, activation='relu', return_sequences=True))
model.add(Dropout(0.3))


model.add(LSTM(units = 80, activation='relu', return_sequences=True))
model.add(Dropout(0.4))


model.add(LSTM(units = 120, activation='relu'))
model.add(Dropout(0.5))


model.add(Dense(units = 1))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss = 'mean_squared_error')
model.fit(x_train, y_train, epochs = 50)
```

```python
model.save('keras_model.h5')
```

```python
data_testing.head()
```

```python
past_100_days = data_training.tail(100)
```

```python
final_df = pd.concat([past_100_days, data_testing], ignore_index=True)
```

```python
final_df.head()
```

```python
input_data = scalar.fit_transform(final_df)
input_data
```

```python
input_data.shape
```

```python
x_test = []
y_test = []

for i in range(100, input_data.shape[0]):
    x_test.append(input_data[i-100: i])
    y_test.append(input_data[i, 0])
```

```python
x_test, y_test = np.array(x_test), np.array(y_test)
print(x_test.shape)
print(y_test.shape)
```

```python
#Making predictions

y_predicted = model.predict(x_test)
```
